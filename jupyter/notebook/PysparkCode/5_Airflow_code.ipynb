```python
from airflow import DAG
from airflow.providers.apache.spark.operators.spark_submit import SparkSubmitOperator
from datetime import datetime

with DAG('bronze_silver_gold_pipeline', #start_date=datetime(2024,1,1), 
         schedule_interval='@daily') as dag:
    bronze = SparkSubmitOperator(task_id='raw_to_bronze', application='/notebook/PysparkCode/Airflow/Bronze_Layer')
    silver = SparkSubmitOperator(task_id='bronze_to_silver', application='/notebook/PysparkCode/Airflow/Silver_Layer')
    gold = SparkSubmitOperator(task_id='silver_to_gold', application='/notebook/PysparkCode/Airflow/Gold_Layer')
    
bronze >> silver >> gold
```
